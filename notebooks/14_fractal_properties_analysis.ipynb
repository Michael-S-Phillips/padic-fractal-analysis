# Fractal Properties Analysis of P-Adic Terrain Embedding

Analyze clustering, density variations, and hierarchical organization in the p-adic embedding space.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import json

cache_dir = Path('/Volumes/Fangorn/padic_fractal_analysis/cache')

# Load metrics
with open(cache_dir / 'fractal_metrics.json') as f:
    metrics = json.load(f)

print("Fractal Metrics Loaded")
```

## Clustering Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

scales = ['l=4', 'l=6']
nn_distances = [metrics[s]['clustering']['mean_nn_distance'] for s in scales]
nn_std = [metrics[s]['clustering']['std_nn_distance'] for s in scales]

ax = axes[0]
ax.bar(scales, nn_distances, yerr=nn_std, capsize=5, color=['blue', 'orange'], alpha=0.7)
ax.set_ylabel('Nearest-Neighbor Distance')
ax.set_title('Clustering: NN Distance Across Scales', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# Density coefficient of variation
density_cv = [metrics[s]['sierpinski_properties']['density_cv'] for s in scales]
ax = axes[1]
ax.bar(scales, density_cv, color=['blue', 'orange'], alpha=0.7)
ax.set_ylabel('Density CV (Clustering Tendency)')
ax.set_title('Sierpinski Properties: Density Variation', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\nClustering Summary:")
for s in scales:
    print(f"\n{s}:")
    print(f"  Mean NN distance: {metrics[s]['clustering']['mean_nn_distance']:.4f}")
    print(f"  Density CV: {metrics[s]['sierpinski_properties']['density_cv']:.4f}")
    print(f"  Clustering strength: {'High' if metrics[s]['sierpinski_properties']['density_cv'] > 2.5 else 'Moderate' if metrics[s]['sierpinski_properties']['density_cv'] > 1.5 else 'Low'}")
```

## Embedding Spread and Geometry

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Embedding span
spans = [metrics[s]['spreading']['embedding_span'] for s in scales]
ax = axes[0]
ax.bar(scales, spans, color=['blue', 'orange'], alpha=0.7)
ax.set_ylabel('Embedding Span (Complex Distance)')
ax.set_title('Embedding Geometric Span', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# Uniformity
uniformity = [metrics[s]['complexity']['embedding_uniformity'] for s in scales]
ax = axes[1]
ax.bar(scales, uniformity, color=['blue', 'orange'], alpha=0.7)
ax.set_ylabel('Uniformity Index')
ax.set_title('Spatial Uniformity (0=uniform, 1=concentrated)', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\nEmbedding Geometry Summary:")
for s in scales:
    print(f"\n{s}:")
    print(f"  Embedding span: {metrics[s]['spreading']['embedding_span']:.4f}")
    print(f"  Spatial uniformity: {metrics[s]['complexity']['embedding_uniformity']:.4f}")
```

## Elevation-Embedding Correlations

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Pearson correlations
corr_real = [metrics[s]['elevation_embedding_correlation']['pearson_real'] for s in scales]
corr_imag = [metrics[s]['elevation_embedding_correlation']['pearson_imag'] for s in scales]

x = np.arange(len(scales))
width = 0.35

ax = axes[0]
ax.bar(x - width/2, corr_real, width, label='Real component', alpha=0.7)
ax.bar(x + width/2, corr_imag, width, label='Imag component', alpha=0.7)
ax.set_ylabel('Pearson Correlation')
ax.set_title('Elevation vs Embedding Correlation', fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(scales)
ax.axhline(0, color='black', linestyle='-', linewidth=0.5)
ax.legend()
ax.grid(axis='y', alpha=0.3)

# Distance matrix correlation
dist_corr = [metrics[s]['elevation_embedding_correlation']['distance_matrix_correlation'] for s in scales]
ax = axes[1]
ax.bar(scales, dist_corr, color=['blue', 'orange'], alpha=0.7)
ax.set_ylabel('Distance Matrix Correlation')
ax.set_title('Elevation Distance vs Embedding Distance', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\nElevation-Embedding Correlation Summary:")
for s in scales:
    print(f"\n{s}:")
    corr = metrics[s]['elevation_embedding_correlation']
    print(f"  Pearson correlation (real): {corr['pearson_real']:.4f}")
    print(f"  Pearson correlation (imag): {corr['pearson_imag']:.4f}")
    print(f"  Distance matrix correlation: {corr['distance_matrix_correlation']:.4f}")
    print(f"  Interpretation: Weak correlation indicates complex hierarchical mapping")
```
