# Fractality Patterns: Deep Dive Analysis

Comprehensive exploration of fractal patterns in p-adic terrain embeddings.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import json

cache_dir = Path('/Volumes/Fangorn/padic_fractal_analysis/cache')

# Load all analysis data
with open(cache_dir / 'fractal_metrics.json') as f:
    metrics = json.load(f)

with open(cache_dir / 'fractality_analysis_report.txt') as f:
    report = f.read()

print("Fractality Pattern Analysis Initialized")
```

## Key Findings Summary

```python
print(report)
```

## Hierarchical Organization Visualization

```python
# Create multi-panel figure showing hierarchical organization
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(3, 2, hspace=0.3, wspace=0.3)

for col, scale_name in enumerate(['l=4', 'l=6']):
    embeddings = np.load(str(cache_dir / f'padic_embeddings_{scale_name}.npy'))
    grid_values = np.load(str(cache_dir / f'dem_grid_{scale_name}.npy'))
    elevations = grid_values.flatten()
    
    # Plot 1: Elevation coloring
    ax = fig.add_subplot(gs[0, col])
    scatter = ax.scatter(embeddings[:, 0], embeddings[:, 1],
                         c=elevations, cmap='terrain', s=100, alpha=0.7)
    ax.set_title(f'{scale_name}: Elevation Mapping', fontweight='bold')
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    plt.colorbar(scatter, ax=ax, label='Elevation (m)')
    
    # Plot 2: Clustering density
    ax = fig.add_subplot(gs[1, col])
    distances_from_origin = np.sqrt(embeddings[:, 0]**2 + embeddings[:, 1]**2)
    scatter = ax.scatter(embeddings[:, 0], embeddings[:, 1],
                         c=distances_from_origin, cmap='cool', s=100, alpha=0.7)
    ax.set_title(f'{scale_name}: Hierarchical Levels', fontweight='bold')
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    plt.colorbar(scatter, ax=ax, label='Radial Distance')
    
    # Plot 3: Elevation distribution
    ax = fig.add_subplot(gs[2, col])
    ax.scatter(elevations, distances_from_origin, alpha=0.6, s=80, edgecolors='black', linewidth=0.5)
    ax.set_xlabel('Elevation (m)')
    ax.set_ylabel('Radial Distance in Embedding')
    ax.set_title(f'{scale_name}: Elevation vs Hierarchical Level', fontweight='bold')
    ax.grid(True, alpha=0.3)

plt.suptitle('Hierarchical Organization of Terrain in P-Adic Space',
             fontsize=14, fontweight='bold')
plt.show()
```

## Metrics Comparison Dashboard

```python
# Create comprehensive metrics comparison
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

scales = ['l=4', 'l=6']
colors = ['#2E86AB', '#A23B72']

# 1. NN Distance
ax = axes[0, 0]
nn_vals = [metrics[s]['clustering']['mean_nn_distance'] for s in scales]
ax.bar(scales, nn_vals, color=colors, alpha=0.7)
ax.set_ylabel('Distance')
ax.set_title('Nearest-Neighbor Distance\n(Clustering)', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# 2. Density CV
ax = axes[0, 1]
cv_vals = [metrics[s]['sierpinski_properties']['density_cv'] for s in scales]
ax.bar(scales, cv_vals, color=colors, alpha=0.7)
ax.set_ylabel('CV')
ax.set_title('Density Clustering Tendency\n(Higher = More Clustered)', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# 3. Embedding Span
ax = axes[0, 2]
span_vals = [metrics[s]['spreading']['embedding_span'] for s in scales]
ax.bar(scales, span_vals, color=colors, alpha=0.7)
ax.set_ylabel('Span')
ax.set_title('Embedding Geometric Span\n(Size in Complex Plane)', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# 4. Elevation Correlation
ax = axes[1, 0]
corr_real = [metrics[s]['elevation_embedding_correlation']['pearson_real'] for s in scales]
corr_imag = [metrics[s]['elevation_embedding_correlation']['pearson_imag'] for s in scales]
x = np.arange(len(scales))
width = 0.35
ax.bar(x - width/2, corr_real, width, label='Real', alpha=0.7)
ax.bar(x + width/2, corr_imag, width, label='Imag', alpha=0.7)
ax.set_ylabel('Correlation')
ax.set_title('Elevation-Embedding Correlation\n(Lower = Less Linear)', fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(scales)
ax.axhline(0, color='black', linestyle='-', linewidth=0.5)
ax.legend()
ax.grid(axis='y', alpha=0.3)

# 5. Distance Matrix Correlation
ax = axes[1, 1]
dist_corr = [metrics[s]['elevation_embedding_correlation']['distance_matrix_correlation'] for s in scales]
ax.bar(scales, dist_corr, color=colors, alpha=0.7)
ax.set_ylabel('Correlation')
ax.set_title('Distance Matrix Correlation\n(Elevation vs Embedding)', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# 6. Summary text
ax = axes[1, 2]
ax.axis('off')
summary_text = f"""KEY METRICS

{scales[0]} → {scales[1]} Scaling:

NN Distance Ratio: {metrics[scales[1]]['clustering']['mean_nn_distance']/metrics[scales[0]]['clustering']['mean_nn_distance']:.4f}
(Lower indicates self-similarity)

Density CV Ratio: {metrics[scales[1]]['sierpinski_properties']['density_cv']/metrics[scales[0]]['sierpinski_properties']['density_cv']:.4f}
(Decreases at finer scales)

Embedding Span Ratio: {metrics[scales[1]]['spreading']['embedding_span']/metrics[scales[0]]['spreading']['embedding_span']:.4f}
(Near 1 = self-similar structure)
"""
ax.text(0.1, 0.5, summary_text, fontsize=11, verticalalignment='center',
       family='monospace', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.suptitle('Comprehensive Metrics Dashboard', fontsize=14, fontweight='bold')
plt.show()
```
